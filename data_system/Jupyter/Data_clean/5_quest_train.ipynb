```python
item_kinds = []
vv = len(items_test)
u = 0
for i in items_test:
    #print(str(i)+" strat")
    u += 1
    if(u % 1000 == 0):
        print("<<<<<"+ str(u) + "<<<<<" + str(u / vv))
    try:
        m = items_map.loc[int(i)]
        item_kinds.append(int(m[0]))
    except:
        print(str(i) + "is not find")
            item_kinds.append(-1)
    
    
```
