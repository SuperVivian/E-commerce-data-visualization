## 循环探究

```python
from  datetime import datetime
from time import time
import numpy as np
import pandas as pd
import gc

rootpath = u"/root/venv/data/jupyter/root/"
```

```python
items = pd.read_csv(rootpath+"ECommAI_ubp_round1_item_feature", delimiter='\t', header=None, names= ["item_id", "class1_id", "class_id", "brand_id", "price"])
gc.collect()
items_id = list(items["item_id"])
```

```python
del items
gc.collect()
```

```python
user = pd.read_csv(rootpath + "finaluser.csv")
user_id = list(user["user_id"])
```

```python
del user
gc.collect()
```

```python
path1 = rootpath + "ECommAI_ubp_round1_train"
u = 5
while(u <= 40):
    data1 = []
    i = -1
    with open(path1, 'r',encoding='gbk',errors='ignore') as f: 
        for line in f:
            i += 1
            if(i < 5000000 * (u - 1)):
                continue
            if(i <= 206676925):
                data1.append(line.split('\t'))
            else:
                break
    print("finish read" + str(u * 5000000))
    data1 = pd.DataFrame(data1,columns=["user_id", "item_id", "behavior_type", "date"])
    data1["date"] = data1["date"].str[0:8]
    data1['user_id'] = data1["user_id"].astype(int)
    start = time()
    data2 = data1.loc[data1['user_id'].isin(user_id)]
    print("finsh delete invailed user")
    data2["item_id"] = data2["item_id"].astype(int)
    data2 = data2.loc[data2['item_id'].isin(items_id)]
    print("finsh delete invailed item")
    print("all time:" + str(time()-start))
    data2.to_csv(rootpath+ str(u * 5000000) +".csv")
    del data1
    del data2
    gc.collect()
    print("<<<<<<<<<<<<<<<<<<<<<<<<<")
    u += 1
```
